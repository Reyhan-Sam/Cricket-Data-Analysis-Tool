**T20 World Cup Cricket Data Pre Processing**

```python
import pandas as pd
import json
```

Process Match Results

```python
with open('json_data/t20_wc_match_results.json') as f:
    data = json.load(f)
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
# Using score card as key ie a unique id( primary)
df_match.rename({'scorecard':'match_id'}, axis = 1, inplace = True)
df_match.head()
```

```python
#creating a match ids dictionary that maps team names to a unique match id
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']

    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']

#match_ids_dict
```

```python
df_match.to_csv('csv_files/dim_match_summary.csv', index = False)
```

Batting Summary

```python
with open('json_data/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])
df_batting = pd.DataFrame(all_records)
df_batting.head(11)
```

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out")
df_batting.head(11)
```

```python
#df_batting.drop(columns = ['dismissal'], inplace = True)
df_batting.tail(20)
```

Cleaning characters for appropriate display

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head(11)
```

```python
df_batting["match_id"] = df_batting["match"].map(match_ids_dict)
df_batting.head(20)
```

```python
df_batting.to_csv("csv_files/fact_batting_summary.csv", index = False)
```

Process Bowling Summary

```python
with open('json_data/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
all_records[:2]
```

```python
df_bowling = pd.DataFrame(all_records)
print(df_bowling.shape)
df_bowling.head()
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids_dict)
df_bowling.head(20)
```

```python
df_bowling.to_csv("csv_files/fact_bowling_summary.csv", index = False)
```
